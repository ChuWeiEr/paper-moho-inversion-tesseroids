```python
%matplotlib inline
from __future__ import division
from fatiando.inversion.base import Misfit
from fatiando.inversion.regularization import Smoothness2D, Damping, CachedMethodPermanent
from fatiando.mesher import Tesseroid
from fatiando.vis import mpl
from fatiando.gravmag import tesseroid
from fatiando import constants, utils, gridder
from scipy.optimize import minimize
import scipy.sparse
import numpy as np
import copy
from joblib import Parallel, delayed
```

```python
class TesseroidRelief(object):
    def __init__(self, area, shape, relief, reference, props=None):
        assert shape[0]*shape[1] == relief.size
        assert len(area) == 4
        assert area[0] < area[1] and area[2] < area[3]
        self.area = area
        self.shape = shape
        x1, x2, y1, y2 = area
        ny, nx = shape
        self.spacing = ((x2 - x1)/nx, (y2 - y1)/ny)
        self._relief = relief
        self.reference = reference
        if props is None:
            self.props = {}
        else:
            self.props = props
        self._i = 0        
    @property
    def relief(self):
        return self._relief
    @relief.setter
    def relief(self, z):
        assert z.size == self.size
        self._relief = z
    @property
    def size(self):
        return relief.size
    def __len__(self):
        return self.size
    def __iter__(self):
        self._i = 0
        return self
    def next(self):
        if self._i >= self.size:
            raise StopIteration
        cell = self.__getitem__(self._i)
        self._i += 1
        return cell
    def __getitem__(self, index):
        ny, nx = self.shape
        x1, x2, y1, y2 = self.area
        dx, dy = self.spacing
        j = index//nx
        i = index - j*nx
        xmin = x1 + i*dx
        xmax = xmin + dx
        ymin = y1 + j*dy
        ymax = ymin + dy
        if self.relief[index] < self.reference:
            zmin = self.relief[index]
            zmax = self.reference
        else:
            zmax = self.relief[index]
            zmin = self.reference
        props = {}
        for p in self.props:
            props[p] = self.props[p][index]
        cell = Tesseroid(xmin, xmax, ymin, ymax, zmax, zmin, props)
        return cell
    def copy(self, deep=False):
        if deep:
            other = copy.deepcopy(self)
        else:
            other = copy.copy(self)
        return other
```

```python
class Moho(Misfit):
    def __init__(self, x, y, z, data, mesh):
        super(Moho, self).__init__(data=data, nparams=mesh.size, islinear=False)
        assert mesh.size == data.size
        self.x = x
        self.y = y
        self.z = z
        self.mesh = mesh.copy(deep=True)
        self.jacobian = CachedMethodPermanent(self, 'jacobian')
        self.hessian = CachedMethodPermanent(self, 'hessian')
        
    def predicted(self, p):
        mesh = self.mesh
        mesh.relief = p
        dens = mesh.props['density']
        should_b_pos = mesh.relief > mesh.reference
        dens[(should_b_pos) & (dens < 0)] *= -1
        dens[(~should_b_pos) & (dens > 0)] *= -1
        mesh.props['density'] = dens
        data = tesseroid.gz(self.x, self.y, self.z, mesh)
        return data
    
    def jacobian(self, p):
        # Tem que ser o modulo
        step = np.abs(utils.si2mgal(2*np.pi*constants.G*self.mesh.props['density']))
        jac = scipy.sparse.diags([step], [0]).tocsr()
        return jac
```

```python
shape = (50, 50)
area = (-40, 40, -40, 40)
lon, lat, h = gridder.regular(area, shape, z=150e3)
dlon, dlat = gridder.spacing(area, shape)
```

```python
modelarea = (area[0] - dlon/2, area[1] + dlon/2, area[2] - dlat/2, area[3] + dlat/2)
f = 0.1
reference = -35e3
relief = 10e3*np.sin(f*lon)*np.cos(f*lat) + reference
```

```python
mpl.pcolor(lon, lat, relief, shape, cmap='Blues_r')
mpl.colorbar()
```

```python
density = 600*np.ones_like(relief)
density[relief < reference] *= -1
mpl.pcolor(lon, lat, density, shape, cmap='Blues_r')
mpl.colorbar()
```

```python
mesh = TesseroidRelief(modelarea, shape, relief, reference, {'density': density})
```

```python
gz = tesseroid.gz(lon, lat, h, mesh)
```

```python
gz = utils.contaminate(gz, 0.5)
```

```python
mpl.contourf(lon, lat, gz, shape, 40, cmap='RdBu_r')
mpl.colorbar()
```

```python
solver = Moho(lon, lat, h, gz, mesh) + 1e-2*Smoothness2D(mesh.shape)
solver.config('levmarq', initial=reference*np.ones(mesh.size)).fit()
```

```python
predicted = solver[0].predicted()
residuals = solver[0].residuals()
```

```python
mpl.figure(figsize=(10, 4))
mpl.subplot(1, 2, 1)
mpl.axis('scaled')
mpl.pcolor(lon, lat, solver.estimate_, shape, cmap='Blues_r')
mpl.colorbar()
mpl.subplot(1, 2, 2)
mpl.axis('scaled')
mpl.pcolor(lon, lat, relief, shape, cmap='Blues_r')
mpl.colorbar()
```

```python
mpl.figure(figsize=(10, 4))
mpl.subplot(1, 2, 1)
mpl.axis('scaled')
mpl.pcolor(lon, lat, predicted, shape, cmap='RdBu_r')
mpl.colorbar()
mpl.subplot(1, 2, 2)
mpl.axis('scaled')
mpl.pcolor(lon, lat, gz, shape, cmap='RdBu_r')
mpl.colorbar()
```
