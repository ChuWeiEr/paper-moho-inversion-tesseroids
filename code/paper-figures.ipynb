The function below takes the data and sets the colorbar limits so that zero falls in the middle of the color range.

```python
def data_minmax(data, shift=0):
    ranges = np.abs([data.min(), data.max()]).max()
    return dict(vmin=-ranges + shift, vmax=ranges + shift)
```

Make a figure of the inversion results.

```python
plt.figure(figsize=(7, 3.5))

cbargs = dict(fraction=0.05, pad=0.1, aspect=40, orientation='horizontal')
tick_locator = ticker.MaxNLocator(nbins=6)

ax = plt.subplot(2, 3, 1)
plt.text(-0.1, 1.03, '(a)', transform=ax.transAxes, **config)
plt.title('Estimated Moho depth (km)', **config)
moho = results['solution'].estimate_    
x, y = bm(moho.lons, moho.lats)
bm.pcolormesh(x, y, -0.001*moho.relief.reshape(moho.shape), cmap='BrBG')
cb = plt.colorbar(**cbargs)
cb.locator = tick_locator
cb.update_ticks()
bm.drawmeridians([-30, 0 , 30], labels=[0, 0, 0, 1], linewidth=0, **config)
bm.drawparallels([20, 40, 60], labels=[1, 0, 0, 0], linewidth=0, **config)

ax = plt.subplot(2, 3, 2)
plt.text(-0.1, 1.03, '(b)', transform=ax.transAxes, **config)
plt.title('True depth minus estimated (mGal)', **config)
diff = -0.001*(model.relief - moho.relief).reshape(moho.shape)
bm.pcolormesh(x, y, diff, cmap='RdYlBu_r', **data_minmax(diff))
cb = plt.colorbar(**cbargs)
cb.locator = tick_locator
cb.update_ticks()
bm.drawmeridians([-30, 0 , 30], labels=[0, 0, 0, 1], linewidth=0, **config)
bm.drawparallels([20, 40, 60], labels=[1, 0, 0, 0], linewidth=0, **config)

ax = plt.subplot(2, 3, 3)
plt.text(-0.1, 1.03, '(c)', transform=ax.transAxes, **config)
plt.title('Residuals (mGal)', **config)
solution = results['solution'][0]
residuals = solution.residuals()
x, y = bm(solution.lon, solution.lat)
bm.contourf(x, y, residuals, 30, tri=True, cmap='RdBu_r', **data_minmax(residuals))
cb = plt.colorbar(**cbargs)
cb.locator = tick_locator
cb.update_ticks()
bm.drawmeridians([-30, 0 , 30], labels=[0, 0, 0, 1], linewidth=0, **config)
bm.drawparallels([20, 40, 60], labels=[1, 0, 0, 0], linewidth=0, **config)

ax = plt.subplot(2, 3, 4)
plt.text(-0.1, 1.03, '(d)', transform=ax.transAxes, **config)
plt.title('Residuals normalized histogram', **config)
plt.text(0.6, 0.8, 
         "mean = {:.2f}\n    std = {:.2f}".format(residuals.mean(), residuals.std()), 
         transform=plt.gca().transAxes, **config)
plt.hist(residuals, bins=15, normed=True, histtype='stepfilled')
plt.xlabel('Residuals (mGal)')
plt.ylabel('Normalized frequency')
ax.set_yticks(ax.get_yticks()[1:-1])

ax = plt.subplot(2, 3, 5)
plt.text(-0.1, 1.03, '(e)', transform=ax.transAxes, **config)
plt.title('Cross-validation curve', **config)
regul_params = results['regul_params']
scores = results['scores']
best = results['best_index']
solver = results['solution']
plt.plot(regul_params, scores, '.-')
plt.plot(regul_params[best], scores[best], '^', 
         color=seaborn.color_palette()[2], 
         label='Minimum ($\mu={:.5f}$)'.format(regul_params[best]))
plt.legend(loc='upper left')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Regularization parameter')
plt.ylabel(u'Mean Square Error (mGal)')
ax.set_yticks(ax.get_yticks()[2:-2])

ax = plt.subplot(2, 3, 6)
plt.text(-0.1, 1.03, '(f)', transform=ax.transAxes, **config)
plt.title("Convergence of Newton's method", **config)
plt.plot(range(solver.stats_['iterations'] + 1), solver.stats_['objective'])
plt.xlabel('Iteration')
plt.yscale('log')
plt.ylabel('Goal function')
ax.set_yticks(ax.get_yticks()[2:-2])

plt.tight_layout(w_pad=0)
plt.savefig('../manuscript/figures/synthetic-simple-results.eps')
```
